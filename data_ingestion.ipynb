```python
import socket
import pandas as pd
import sqlalchemy as sa
from sqlalchemy.engine import URL
from sqlalchemy import create_engine
 
from sqlalchemy import create_engine, MetaData, Table, select, Column, BigInteger, Numeric, String

pc_name= socket.gethostname()
 
 
def test_connection():
 
    try:
 
 
        database_name='python'
        pc_name= socket.gethostname()
 
        # Create a connection string
        connection_string = "DRIVER={SQL Server};SERVER="+pc_name+"\SQLEXPRESS;DATABASE="+database_name+";"

        # Establish a connection
        connection_url = URL.create("mssql+pyodbc", query={"odbc_connect": connection_string})
        conn_string=str(connection_url)
        print("Connection SUCCESSFULL")
 
        #SELECT 
        
        query="SELECT* from dbo.db_1;"
        engine = create_engine(conn_string)
        df_mpt_profile = pd.read_sql_query(sa.text(query), conn_string)
        print("Original DataFrame:")
        print(df_mpt_profile)

        # Update query 1
        update_query = "UPDATE dbo.db_1 SET Cumulative_cases = 300 WHERE Country = 'Albania' AND Date = '22-01-2020';"

        # Update query 2
        update_query_2 = "UPDATE dbo.db_1 SET Cumulative_cases = 500 WHERE Country = 'Albania' AND Date = '23-01-2020';"

        # Update query 3
        update_query_3 = "UPDATE dbo.db_1 SET Cumulative_cases = 700 WHERE Country = 'Albania' AND Date = '25-01-2020';"
        
        # Create a new connection for the update
        update_engine = create_engine(conn_string)
        with update_engine.connect() as connection:
            
                # Execute Update Query 1
                connection.execute(sa.text(update_query))
                connection.commit()

                # Execute Update Query 2
                connection.execute(sa.text(update_query_2))
                connection.commit()

                # Execute Update Query 3
                connection.execute(sa.text(update_query_3))
                connection.commit()
        
                # Fetch the data again after the update
                df_mpt_profile_updated = pd.read_sql_query(sa.text(query), conn_string)
        
                print("Updated DataFrame:")
                print(df_mpt_profile_updated)
                print("Update Query Successful")

            # Filter Query 1
                filter_query_1 = "SELECT * FROM dbo.db_1 WHERE Country = 'Albania';"
                df_filtered_1 = pd.read_sql_query(sa.text(filter_query_1), conn_string)
                print("Filtered DataFrame 1:")
                print(df_filtered_1)

                # Filter Query 2
                filter_query_2 = "SELECT Daily_death FROM dbo.db_1 WHERE Country = 'Albania' AND Date = '22-01-2020';"
                df_filtered_2 = pd.read_sql_query(sa.text(filter_query_2), conn_string)
                print("Filtered DataFrame 2:")
                print(df_filtered_2)

            # ALter Query

                alter_query = "ALTER TABLE dbo.db_1 ADD Total_deaths INT, New_column INT;"
                connection.execute(sa.text(alter_query))
                connection.commit()
                df_altered = pd.read_sql_query(sa.text(query), conn_string)
                print("Table Altered Successfully")
                print(df_altered)

            # Delete Query
            
                delete_query = "DELETE FROM dbo.db_1 WHERE Country = 'France';"
                connection.execute(sa.text(delete_query))
                connection.commit()
                df_deleted = pd.read_sql_query(sa.text(query), conn_string)
                print(df_deleted)

             #Specify the columns to be removed
            
                columns_to_remove = ['New_column']
    
            # Construct the ALTER TABLE query to drop multiple columns
            
                alter_query1 = "ALTER TABLE dbo.db_1 DROP COLUMN " + ", ".join(columns_to_remove) + ";"
                connection.execute(sa.text(alter_query1))
                connection.commit()
                df_altered1 = pd.read_sql_query(sa.text(query), conn_string)
                print("Table Altered Successfully")
                print(df_altered1)
                    
    
    except Exception as e:
        print("Error:", str(e))

 
test_connection()
```
