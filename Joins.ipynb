```python
import pandas as pd
import numpy as np
import warnings

csv_file_path = './merged_data.csv'

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    df = pd.read_csv(csv_file_path)

# Display the DataFrame
df.head(100)
```

```python
df2 = pd.read_csv('./archive/India_COVID19.csv')

df2.head()
```

INNER JOIN

```python
inner_join_data = pd.merge(df, df2, on='Country', how='inner', suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')

print(inner_join_data.head())
```

OUTER JOIN

```python
inner_join_data = pd.merge(df, df2, on='Country', how='outer', suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')

print(inner_join_data.head(500))
```

LEFT JOIN

```python
inner_join_data = pd.merge(df, df2, on='Country', how='left', suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')

print(inner_join_data.head())
```

RIGHT JOIN

```python
inner_join_data = pd.merge(df, df2, on='Country', how='right', suffixes=('', '_drop')).filter(regex='^(?!.*_drop)')

print(inner_join_data.head())
```

PARTIONING

```python
# Partitioning based on the 'Country' column
partitioned_data = df.groupby('Country')

# Display the partitions
for country, group in partitioned_data:
    print(f"Country: {country}")
    print(group)
    print('\n')
```

INDEXING

```python
import pandas as pd
import numpy as np
import warnings

csv_file_path = './merged_data.csv'

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    data = pd.read_csv(csv_file_path)
    
# Set 'Country' and 'Date' as multi-level index
data.set_index(['Date'], inplace=True)

# Display the DataFrame with the new index
print(data.head())
```

```python
print(data.iloc[500])   # Select row at index position 1
```

```python
print(data.iloc[:, 3])   # Select column at index position 0
```

PRIMARY KEY

```python
import pandas as pd
import numpy as np
import warnings

csv_file_path = './merged_data.csv'

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    data = pd.read_csv(csv_file_path)
    
# Set the combination of 'Date', 'State', and 'Country' as the composite primary key
data.set_index(['Date', 'State', 'Country'], inplace=True)

# Display the DataFrame with the new index
print(data.head(300))
```
